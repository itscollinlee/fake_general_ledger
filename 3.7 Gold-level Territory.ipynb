```python
%sql
/**************************************
Create Gold Table With Metadata Columns
***************************************/
CREATE OR REPLACE TABLE fake_general_ledger.default.dim_territory_mt AS

SELECT
    territory_key,
    country,
    region,
    city,
    'GL' AS record_source,
    batch_id,
    CURRENT_TIMESTAMP() AS gold_processed_timestamp

FROM fake_general_ledger.default.silver_territory
```

```python
%sql
/***************************************
Create Gold View Without Metadata Columns
****************************************/
CREATE OR REPLACE VIEW fake_general_ledger.default.dim_territory_vw AS

SELECT
    territory_key,
    country,
    region,
    city

FROM fake_general_ledger.default.dim_territory_mt
```
