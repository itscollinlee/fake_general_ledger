```python
%sql
/**************************************
Create Gold Table With Metadata Columns
***************************************/
CREATE OR REPLACE TABLE fake_general_ledger.default.cash_flow_statement_mt AS

SELECT
    cf.type AS statement_type,
    cf.subtype AS statement_subtype,
    SUM(gl.amount) AS amount
FROM fake_general_ledger.default.silver_general_ledger gl

INNER JOIN fake_general_ledger.default.silver_statement_of_cash_flow cf
    ON gl.account_key = cf.account_key

GROUP BY
    cf.type,
    cf.subtype;
```

```python
%sql
/***************************************
Create Gold View Without Metadata Columns
****************************************/
CREATE OR REPLACE VIEW fake_general_ledger.default.cash_flow_statement_vw AS

SELECT
    statement_type,
    statement_subtype,
    amount
FROM fake_general_ledger.default.cash_flow_statement_mt;
```
