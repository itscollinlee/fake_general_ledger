```python
%sql
/**************************************
Create Gold View Without Metadata Columns
***************************************/
CREATE OR REPLACE TABLE fake_general_ledger.default.account_summary_vw AS

SELECT
    account_level_1,
    account_level_2,
    class,
    subclass_level_1,
    subclass_level_2,
    SUM(amount) AS total_amount
FROM fake_general_ledger.default.fact_general_ledger_mt
GROUP BY
    account_level_1,
    account_level_2,
    class,
    subclass_level_1,
    subclass_level_2;
```
