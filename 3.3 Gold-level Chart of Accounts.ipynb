```python
%sql
/**************************************
Create Gold Table With Metadata Columns
***************************************/
CREATE OR REPLACE TABLE fake_general_ledger.default.dim_gl_account_mt AS

SELECT
    account_key,
    account AS account_level_1,
    subaccount AS account_level_2,
    class,
    subclass AS subclass_level_1,
    subclass2 AS subclass_level_2,
    report,
    'GL' AS record_source,
    batch_id,
    CURRENT_TIMESTAMP() AS gold_processed_timestamp

FROM fake_general_ledger.default.silver_chart_of_accounts;
```

```python
%sql
/***************************************
Create Gold View Without Metadata Columns
****************************************/
CREATE OR REPLACE VIEW fake_general_ledger.default.dim_gl_account_vw AS

SELECT
    account_key,
    account_level_1,
    account_level_2,
    class,
    subclass_level_1,
    subclass_level_2,
    report

FROM fake_general_ledger.default.dim_gl_account_mt;
```
