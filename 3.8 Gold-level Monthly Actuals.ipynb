```python
%sql
/**************************************
Create Gold View Without Metadata Columns
***************************************/
CREATE OR REPLACE VIEW fake_general_ledger.default.monthly_actuals_vw AS

SELECT
    year,
    month,
    account_level_1,
    account_level_2,
    country,
    region,
    SUM(amount) AS actual_amount
FROM fake_general_ledger.default.fact_general_ledger_mt
GROUP BY
    year,
    month,
    account_level_1,
    account_level_2,
    country,
    region;
```
