```python
%sql
/**************************************
Create Gold Table With Metadata Columns
***************************************/
CREATE OR REPLACE TABLE fake_general_ledger.default.fact_general_ledger_mt AS

SELECT
    gl.entryno,
    gl.date AS transaction_date,
    cal.year,
    cal.quarter,
    cal.month,
    gl.account_key,
    coa.account_level_1,
    coa.account_level_2,
    coa.class,
    coa.subclass_level_1,
    coa.subclass_level_2,
    gl.territory_key,
    terr.country,
    terr.region,
    terr.city,
    gl.amount,
    gl.record_ingestion_timestamp,
    'GL' AS record_source,
    gl.batch_id,
    CURRENT_TIMESTAMP() AS gold_processed_timestamp

FROM fake_general_ledger.default.silver_general_ledger gl

LEFT JOIN fake_general_ledger.default.dim_gl_account_mt coa
    ON gl.account_key = coa.account_key
LEFT JOIN fake_general_ledger.default.dim_territory_mt terr
    ON gl.territory_key = terr.territory_key
LEFT JOIN fake_general_ledger.default.dim_calendar_mt cal
    ON gl.date = cal.calendar_date;
```

```python
%sql
/***************************************
Create Gold View Without Metadata Columns
****************************************/
CREATE OR REPLACE VIEW fake_general_ledger.default.fact_general_ledger_vw  AS

SELECT
    gl.entryno,
    gl.transaction_date,
    gl.year,
    gl.quarter,
    gl.month,
    gl.account_key,
    gl.account_level_1,
    gl.account_level_2,
    gl.class,
    gl.subclass_level_1,
    gl.subclass_level_2,
    gl.territory_key,
    gl.country,
    gl.region,
    gl.city,
    gl.amount

FROM fake_general_ledger.default.fact_general_ledger_mt gl;
```
