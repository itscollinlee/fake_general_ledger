```python
%sql
/**************************************
Create Gold Table With Metadata Columns
***************************************/

CREATE OR REPLACE TABLE fake_general_ledger.default.dim_calendar_mt AS

SELECT
    date AS calendar_date,
    year,
    quarter,
    month,
    day,
    'GL' AS record_source,
    batch_id,
    CURRENT_TIMESTAMP() AS gold_processed_timestamp

FROM fake_general_ledger.default.silver_level_calendar;
```

```python
%sql
/***************************************
Create Gold View Without Metadata Columns
****************************************/
CREATE OR REPLACE VIEW fake_general_ledger.default.dim_calendar_vw  AS

SELECT
    calendar_date,
    year,
    quarter,
    month,
    day

FROM fake_general_ledger.default.dim_calendar_mt;
```
