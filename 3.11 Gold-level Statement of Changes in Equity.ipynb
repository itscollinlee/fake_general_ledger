```python
%sql
/**************************************
Create Gold Table With Metadata Columns
***************************************/
CREATE OR REPLACE TABLE fake_general_ledger.default.statement_of_changes_in_equity_mt AS

SELECT
    eq.type AS statement_type,
    eq.balancetype AS balance_type,
    SUM(gl.amount) AS amount,
    'GL' AS record_source,
    MAX(gl.batch_id) AS batch_id,
    CURRENT_TIMESTAMP() AS gold_processed_timestamp

FROM fake_general_ledger.default.silver_general_ledger gl

INNER JOIN fake_general_ledger.default.silver_statement_of_changes_in_equity eq
    ON gl.account_key = eq.account_key

GROUP BY
    eq.type,
    eq.balancetype;
```

```python
%sql
/***************************************
Create Gold View Without Metadata Columns
****************************************/
CREATE OR REPLACE VIEW fake_general_ledger.default.statement_of_changes_in_equity_vw AS

SELECT
    statement_type,
    balance_type,
    amount
FROM fake_general_ledger.default.statement_of_changes_in_equity_mt;
```
