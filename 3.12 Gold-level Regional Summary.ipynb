```python
%sql
/**************************************
Create Gold View Without Metadata Columns
***************************************/
CREATE OR REPLACE TABLE fake_general_ledger.default.region_summary_vw AS

SELECT
    country,
    region,
    city,
    SUM(amount) AS total_amount
FROM fake_general_ledger.default.fact_general_ledger_mt

GROUP BY
    country,
    region,
    city;
```
