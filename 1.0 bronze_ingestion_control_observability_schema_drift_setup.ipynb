```python
# ---------------------------------------------------------
# 0. Run Imports 
# ---------------------------------------------------------

import os
from pyspark.sql import SparkSession
from pyspark.sql import functions as F
from pyspark.sql.types import *
from delta.tables import DeltaTable
import hashlib
```

```python
# ---------------------------------------------------------
# 1. Set-up Empty Ingestion Control Table
# ---------------------------------------------------------

spark = SparkSession.builder.appName("ingestion_control_tbl").getOrCreate()

schema = StructType([
    StructField("source_system", StringType(), True),
    StructField("bronze_table", StringType(), True),
    StructField("file_path", StringType(), True),
    StructField("ingestion_status", StringType(), True),
    StructField("pipeline_run_id", StringType(), True),
    StructField('batch_id', StringType(), True),
    StructField('file_hash', StringType(), True),
    StructField('file_size', LongType(), True),
    StructField("ingestion_timestamp", TimestampType(), True),
    StructField("completion_timestamp", StringType(), True),
    StructField("row_count", LongType(), True),
    StructField('error_message', StringType(), True)
])

empty_df = spark.createDataFrame([], schema)
empty_df.write.mode("overwrite").saveAsTable("fake_general_ledger.default.bronze_ingestion_control")
```

```python
# ---------------------------------------------------------
# 2. Set-up Empty Ingestion Schema Drift Table
# ---------------------------------------------------------

spark = SparkSession.builder.appName("ingestion_schema_drift_tbl").getOrCreate()

schema = StructType([
    StructField("source_system", StringType(), True),
    StructField("bronze_table", StringType(), True),
    StructField('batch_id', StringType(), True),
    StructField("pipeline_run_id", StringType(), True),
    StructField("file_path", StringType(), True),
    StructField("change_type", StringType(), True),
    StructField("column_name", StringType(), True),
    StructField("old_data_type", StringType(), True),
    StructField("new_data_type", StringType(), True),
    StructField("detected_timestamp", TimestampType(), True)
])

empty_df = spark.createDataFrame([], schema)
empty_df.write.mode("overwrite").saveAsTable("fake_general_ledger.default.bronze_ingestion_schema_drift")
```
